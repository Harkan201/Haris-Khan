```python
import random
import random
from collections import Counter

def hat_problem():
    # Initialize the hat with numbers 1 through 10
    hat = list(range(1, 11))

    # Repeat the process until only one number remains
    while len(hat) > 1:
        # Draw two numbers at random from the hat
        x, y = random.sample(hat, 2)
        # Calculate the difference
        diff = abs(x - y)
        # Remove the drawn numbers from the hat
        hat.remove(x)
        hat.remove(y)
        # Add the difference to the hat
        hat.append(diff)

    # Return the final remaining number
    return hat[0]

# Run the simulation
result = hat_problem()
print("The final remaining number is:", result)

# Run the simulation 100 times
results = []
for _ in range(100):
    result = hat_problem()
    results.append(result)

# Print all the results
print("Results of 100 simulations:")
print(results)
occurrences = Counter(results)

# Print the occurrences of each number
print("Occurrences of each number:")
for number, count in occurrences.items():
    print(f"Number {number}: {count} times")
```
